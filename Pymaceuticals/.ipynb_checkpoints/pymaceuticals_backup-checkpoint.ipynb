```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')
```

```python
# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data

mouse_df = pd.read_csv(mouse_drug_data_to_load)

clinical_df = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset

merged_clinical_mouse = pd.merge(clinical_df, mouse_df, how ='outer', on='Mouse ID')

# Display the data table for preview

merged_clinical_mouse.head(20)
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
tumor_vol_mean = merged_clinical_mouse.groupby(["Drug", "Timepoint"]).mean()

#reset index
tumor_vol_mean_reset = tumor_vol_mean.reset_index()

# Convert to DataFrame
tumor_vol_mean_df = pd.DataFrame(data=tumor_vol_mean_reset)

# Preview DataFrame

tumor_vol_mean_df.head(10)
```

```python
# Use group by fuction to group by "drug" and "timepoint" and aggregate by standard error

stnd_error = merged_clinical_mouse.groupby(["Drug", "Timepoint"]).sem()

# stnd_error = stnd_error.reset_index()

# drop Mouse ID column since it comes in as NaN after aggragating

stnd_error = stnd_error.dropna(axis=1)
stnd_error = stnd_error.rename(columns={'Metastatic Sites':'Metastatic Sites Stnd Error','Tumor Volume (mm3)':'Volume Standard Error'})

# preview dataframe
stnd_error.head(20)
```

```python
test = pd.merge(tumor_vol_mean_df, stnd_error, on=['Drug','Timepoint'], how='outer')

test = test.set_index(['Drug', 'Timepoint'])

test.head()
```

```python
test = test.unstack().T

test
```

```python
# Minor Data Munging to Re-Format the Data Frames

# test2 = merged_clinical_mouse.groupby(['Drug','Timepoint']).mean()

reformat_volume = tumor_vol_mean.drop(columns=['Metastatic Sites'])

reformat_volume = reformat_volume.unstack().T

reformat_volume = reformat_volume.reset_index()
reformat_volume = reformat_volume.drop(columns='level_0')
reformat_volume = reformat_volume.set_index(['Timepoint'])

# Preview that reformatting worked
reformat_volume
```

```python
reformat_se = stnd_error.drop(columns=["Metastatic Sites Stnd Error"])

reformat_se = reformat_se.unstack().T
reformat_se = reformat_se.reset_index()
reformat_se = reformat_se.drop(columns='level_0')
reformat_se = reformat_se.set_index('Timepoint')

reformat_se.head()
```

```python
x_axis = reformat_volume.index.values

print(x_axis)
```
